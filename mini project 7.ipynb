```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso, Ridge


rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)


model = LinearRegression()
model.fit(x.reshape(-1, 1), y)
y_pred = model.predict(x.reshape(-1, 1))


plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.title('Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Print the coefficients and intercept
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
model_lasso = Lasso(alpha=0.1)
model_lasso.fit(x.reshape(-1, 1), y)
y_pred_lasso = model_lasso.predict(x.reshape(-1, 1))

model_ridge = Ridge(alpha=0.1)
model_ridge.fit(x.reshape(-1, 1), y)
y_pred_ridge = model_ridge.predict(x.reshape(-1, 1))

# Plot the data and the lines of best fit
plt.scatter(x, y)
plt.plot(x, y_pred, color='red', label='Linear Regression')
plt.plot(x, y_pred_lasso, color='green', label='Lasso Regression')
plt.plot(x, y_pred_ridge, color='orange', label='Ridge Regression')
plt.title('Linear Regression vs Lasso vs Ridge')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Print the coefficients and intercept for each model
print('Linear Regression Coefficients:', model.coef_)
print('Linear Regression Intercept:', model.intercept_)

print('Lasso Regression Coefficients:', model_lasso.coef_)
print('Lasso Regression Intercept:', model_lasso.intercept_)

print('Ridge Regression Coefficients:', model_ridge.coef_)
print('Ridge Regression Intercept:', model_ridge.intercept_)
```
